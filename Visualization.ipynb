```python
#Imports

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read Data

data = pd.read_csv("data/plant_data_collected.csv")
```

```python
data["HumidityRequirement"] = pd.Categorical(data["HumidityRequirement"], ordered=True)
data["GrowingSeason"] = pd.Categorical(data["GrowingSeason"], ordered=True)
data["FloweringPeriod"] = data["FloweringPeriod"].astype("string")
data["Benefit"] = data["Benefit"].astype("string")
data["Notes"] = data["Notes"].astype("string")
data["Usability"] = data["Usability"].astype("string")
data["Family"] = pd.Categorical(data["Family"], ordered=True)
data["SuitableZone"] = pd.Categorical(data["SuitableZone"], ordered=True)
data["Location"] = pd.Categorical(data["Location"], ordered=True)
data["Soil"] = pd.Categorical(data["Soil"], ordered=True)
data["Height"] = data["Height"].astype("string")
data["Herkunft"] = data["Herkunft"].astype("string")
data["Images"] = data["Images"].astype("string")
data["Name"] = data["Name"].astype("string")
data["BotanicName"] = data["BotanicName"].astype("string")
data["FlowerColor"] = data["FlowerColor"].astype("string")
data.dtypes
```
